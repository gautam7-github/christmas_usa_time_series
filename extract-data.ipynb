```python
URL = "https://explodingtopics.com/blog/christmas-spending-stats"
```

```python
import pandas as pd
import requests as rq
```

```python
webpage = rq.get(URL)._content
```

```python
dfs = pd.read_html(webpage.decode())
```

```python
len(dfs)
```

```python
dfs[1]
```

```python
for idx, df in enumerate(dfs):
    df.to_csv(f"data-tables/{idx}.csv",header=False,index=False)
```
